```python
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'] = df['Age'].fillna(df['Age'].mean())  # Perubahan di sini
df['Salary'] = df['Salary'].fillna(df['Salary'].median())  # Perubahan di sini
df.dropna(subset=['Name'], inplace=True)  # Ini tetap bisa digunakan
print('After cleaning:\n', df)
```

```python
# Exercise 2: Standardizing Categorical Data
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Memuat dataset Iris dari URL
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, header=None, names=columns)

# Menampilkan beberapa baris pertama dari dataset
print("Dataframe asli:")
print(df.head())

# Mencari nilai yang hilang
nilai_hilang = df.isnull().sum()
print("\nNilai yang hilang di setiap kolom:")
print(nilai_hilang[nilai_hilang > 0])

# Normalisasi kolom numerik
kolom_numerik = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
scaler = MinMaxScaler()
df[kolom_numerik] = scaler.fit_transform(df[kolom_numerik])
print("\nKolom numerik setelah dinormalisasi:")
print(df[kolom_numerik].head())

# Menstandarisasi kolom kategorikal
kolom_kategorikal = df.select_dtypes(include=['object']).columns.tolist()
for col in kolom_kategorikal:
    df[col] = df[col].str.lower().str.strip()
print("\nKolom kategorikal setelah distandarisasi:")
print(df[kolom_kategorikal].head())

# Menghapus duplikat
df = df.drop_duplicates()
print("\nDataframe setelah menghapus duplikat:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Q1', 'Q2', 'Q3', 'Q4']
sales_person1 = [100, 150, 200, 250]
sales_person2 = [120, 180, 240, 300]
sales_person3 = [90, 130, 250, 270]

bar_width = 0.25
x = np.arange(len(categories))  # Lokasi kategori

# Membangun multi-bar chart
plt.figure(figsize=(10, 6))
plt.bar(x, sales_person1, width=bar_width, label='Sales Person 1', color='b', align='center')
plt.bar(x + bar_width, sales_person2, width=bar_width, label='Sales Person 2', color='g', align='center')
plt.bar(x + 2 * bar_width, sales_person3, width=bar_width, label='Sales Person 3', color='r', align='center')

# Menambahkan judul dan label
plt.title('Sales Comparison by Quarter')
plt.xlabel('Quarters')
plt.ylabel('Sales ($)')
plt.xticks(x + bar_width, categories)
plt.legend()

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd              # Mengimpor library pandas untuk manipulasi data
import numpy as np               # Mengimpor library numpy untuk operasi numerik
from scipy import stats          # Mengimpor library stats dari scipy untuk perhitungan statistik
from sklearn.preprocessing import MinMaxScaler  # Mengimpor MinMaxScaler untuk normalisasi data

# 1. Mengimpor dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"  # URL dataset
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']  # Nama kolom dataset
df = pd.read_csv(url, names=column_names)  # Membaca dataset dan memberikan nama kolom

# 2. Melihat data awal
print("Data Awal:")
print(df.head())  # Menampilkan 5 baris pertama dari dataset

# 3. Menghapus nilai yang hilang jika ada
df = df.dropna()  # Menghapus baris yang memiliki nilai hilang (NaN)

# 4. Menemukan deskripsi statistik untuk identifikasi outlier
print("\nDeskripsi Statistik:")
print(df.describe())  # Menampilkan ringkasan statistik dari dataset

# 5. Menghitung Z-score untuk mendeteksi outlier
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))  # Menghitung Z-score untuk fitur numerik
df = df[(z_scores < 3).all(axis=1)]  # Membuang baris dengan Z-score ≥ 3, sehingga menghapus outlier

# 6. Normalisasi: Menggunakan Min-Max Scaler untuk mengubah skala data ke rentang [0, 1]
scaler = MinMaxScaler()  # Membuat objek MinMaxScaler
df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']] = scaler.fit_transform(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])  # Normalisasi fitur

# 7. Menampilkan data setelah normalisasi
print("\nData Setelah Normalisasi:")
print(df.head())  # Menampilkan 5 baris pertama dari data yang sudah dinormalisasi
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Memuat dataset Iris
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv'
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df = pd.read_csv(url, names=column_names)

# 2. Visualisasi 1: Distribusi Bunga Berdasarkan Spesies
plt.figure(figsize=(8, 5))
sns.countplot(x='species', data=df, palette='pastel')
plt.title('Distribusi Bunga Berdasarkan Spesies')
plt.xlabel('Spesies Bunga')
plt.ylabel('Jumlah Bunga')
plt.show()

# 3. Visualisasi 2: Distribusi Panjang Sepal
plt.figure(figsize=(8, 5))
sns.histplot(df['sepal_length'], bins=20, kde=True, color='skyblue')
plt.title('Distribusi Panjang Sepal')
plt.xlabel('Panjang Sepal (cm)')
plt.ylabel('Frekuensi')
plt.show()

# 4. Visualisasi 3: Korelasi Antara Panjang dan Lebar Petal
plt.figure(figsize=(8, 5))
sns.scatterplot(x='petal_length', y='petal_width', data=df, hue='species', palette='coolwarm', alpha=0.7)
plt.title('Korelasi Antara Panjang dan Lebar Petal')
plt.xlabel('Panjang Petal (cm)')
plt.ylabel('Lebar Petal (cm)')
plt.legend(title='Spesies', loc='upper right')
plt.show()

# 5. Visualisasi 4: Peluang Berdasarkan Spesies (Rata-rata Panjang dan Lebar)
plt.figure(figsize=(8, 5))
sns.barplot(x='species', y='petal_length', data=df, palette='viridis', estimator='mean')
plt.title('Peluang Berdasarkan Spesies (Rata-rata Panjang Petal)')
plt.xlabel('Spesies Bunga')
plt.ylabel('Rata-rata Panjang Petal (cm)')
plt.show()
```
